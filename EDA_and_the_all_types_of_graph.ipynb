```python
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
dataset = [10,25,14,26,35,45,67,90,40,50,60,10,16,18,20]
# calculate the skewness
print(skew(dataset, axis=0,bias=True))

#calculate the kurtosis
print(kurtosis(dataset, axis=0,bias=True))

```

```python
x:
scipy.stats.skew(array,axis=0,bias =True)

Parameters:
array:it represents the input array(or object) containing elements. axis: It signifies the axis along which we want to find the skewness ValueError
 (By deafault axis =0).bias =False: calculations are  correlated to ststical bias

  Return type:
  Skewness value of the dataset, along with the axis
```

```python
import numpy as np
x= np.array([1,3,5,7,8,9,10,15])
y =np.array([10,20,30,40,50,60,70,80])
```

```python
def Person_correlation(X,Y):
    if len(X)==len(Y):
        Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
        Sum_x_squared = sum((X-X.mean())**2)
        Sum_y_squared = sum((Y-Y.mean())**2)
        corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
    return  corr

print(Person_correlation(x,y))
print(Person_correlation(x,x))
print(np.corrcoef(x,y))

```
